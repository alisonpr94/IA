```
# 1º Contar o número de reprovações de cada aluno
# 2º Após contar o número de reprovações, pode remover as linhas, pois não serão mais necessárias
# 3º Fazer a média das disciplinas de cada aluno dos dois primeiros períodos
# 4º Após realizar as médias de cada aluno, reduzir o dataset para uma linha por aluno, no caso, teremos 170 linhas
# 5º Anexar no final duas colunas: coluna com número de reprovações e coluna com as médias

''' OBS.: podem ser duas médias: (1) média 1 pertence as cinco disciplinas do primeiro semestre e média 2 pertence as cinco
disciplinas do segundo semestre, no caso teremos 3 colunas para anexar; (2) a segunda opção é fazer a média com as 10 disciplinas e
anexar apenas duas colunas. Tudo depende de como vai ficar nosso modelo no final
'''
```

```
# Número de alunos que se formaram
lista_id = [] # Lista que guarda os id's eliminando repetições
for i in data.id:
    if i not in lista_id:
        lista_id.append(i)
len(lista_id)
```

```
# Lista que guarda o número de reprovações de cada aluno
# Cada posição da lista corresponde a um aluno graduado

lista_num_reprovacoes = []

for i in lista_id:
    df = data[data.id == i] # Filtra as disciplinas de cada aluno
    df = df[df.situacao_discente_componente != 'AP'] # Filta as reprovações
    numero_rep = df.shape[0] # Só vai restar as reprovações, o número de linhas é a quantidade de reprovações
    lista_num_reprovacoes.append(numero_rep)
    
```

### **Como eu disse, nem todos os graduados tem registrado todas as disciplinas aprovadas, por isso alguns tem média 0.**

```
disciplinas = ['PROGRAMAÇÃO DE COMPUTADORES 1 CN', 'INTRODUÇÃO À COMPUTAÇÃO', 'CÁLCULO 1', 'MATEMÁTICA DISCRETA', 'LÓGICA MATEMÁTICA',
               'PROGRAMAÇÃO DE COMPUTADORES 2 CN', 'ÁLGEBRA LINEAR', 'FÍSICA PARA COMPUTAÇÃO', 'SISTEMAS DIGITAIS', 'CÁLCULO 2']

medias = []
lista_com_ids = lista_id

for i in range(170):
    id = lista_com_ids.pop(0)
    df = data[data.id == id] # Filtra os dados do aluno i
    df = df[df.situacao_discente_componente == 'AP'] # Filtra a situação de aprovações do aluno i
    materias = list(df.nome_componente) # Cria uma lista com todas as disciplinas aprovadas
    notas = []
    
    # Busca pelas 10 primeiras disciplinas do curso
    for disc in disciplinas:
        # Estranhamente não aparece todas as disciplinas dos graduados, então eu verifico se ela existe antes de pegar as "10" notas
        if disc not in materias:
            continue
        else:   
            d = df[df.nome_componente == disc]
            notas.append(d.loc[d.index, 'nota_discente_componente'].get_values()[0])
    if sum(notas) != 0:
        medias.append(sum(notas)/len(notas))
    else:
        medias.append(sum(notas))
    
medias[:22]
```
