**Descrição da Rede Neural**

 - A rede neural artificial possui a camada de entrada, um número **N** de camadas intermediárias e uma camada de saída;
 - A rede realiza **forward** e **backward** no treinamento;
 - Os pesos são gerados **aleatoriamente** para cada neurônio em suas respectivas camadas;
 - A função de ativação utilizada é a **Sigmoide**;
 - Os pesos são atualizados de acordo com a descida de gradiente por meio da regra **delta**;
 - Os dados não foram simplemente dividos em conjuntos de treinamento e teste, mas sim aplicou-se a **validação cruzada** com 10 folds, pois assim aproveita-se melhor os dados e testa a capacidade de generalização da rede.
  - Os folds foram montados misturando-se alunos graduados e não graduados, os graduados que cumpriram as regras impostas por nos para avaliação de desempenho, foram misturados com alunos graduados que não cumpriram e com alunos não graduados.
  - As regras impostas são média global maior do que seis ao término do curso.

```
class Preprocessamento:
    def __init__(self, data):
        self.lista_id = []
        self.lista_num_reprovacoes = []
        self.medias = []
        self.classes = []
        self.data = data
        
    def captura_ids_alunos(self):
        # Número de alunos que se formaram
        self.lista_id = [] # Lista que guarda os id's eliminando repetições
        for i in self.data.id:
            if i not in self.lista_id:
                self.lista_id.append(i)
    
    def numero_de_reprovacoes(self):
        # Lista que guarda o número de reprovações de cada aluno
        # Cada posição da lista corresponde a um aluno graduado

        for i in self.lista_id:
            df = self.data[data.id == i] # Filtra as disciplinas de cada aluno
            df = df[df.situacao_discente_componente != 'AP'] # Filta as reprovações
            numero_rep = df.shape[0] # Só vai restar as reprovações, o número de linhas é a quantidade de reprovações
            self.lista_num_reprovacoes.append(numero_rep)
            
    def medias_primeiras_disciplinas(self):
        disciplinas = ['PROGRAMAÇÃO DE COMPUTADORES 1 CN', 'INTRODUÇÃO À COMPUTAÇÃO', 'CÁLCULO 1', 'MATEMÁTICA DISCRETA', 'LÓGICA MATEMÁTICA',
               'PROGRAMAÇÃO DE COMPUTADORES 2 CN', 'ÁLGEBRA LINEAR', 'FÍSICA PARA COMPUTAÇÃO', 'SISTEMAS DIGITAIS', 'CÁLCULO 2']

        self.medias = []
        lista_com_ids = self.lista_id

        for i in range(len(self.lista_id)):
            id = lista_com_ids.pop(0)
            df = self.data[self.data.id == id] # Filtra os dados do aluno i
            df = df[df.situacao_discente_componente == 'AP'] # Filtra a situação de aprovações do aluno i
            materias = list(df.nome_componente) # Cria uma lista com todas as disciplinas aprovadas
            notas = []

            # Busca pelas 10 primeiras disciplinas do curso
            for disc in disciplinas:
                # Estranhamente não aparece todas as disciplinas dos graduados, então eu verifico se ela existe antes de pegar as "10" notas
                if disc not in materias:
                    continue
                else:   
                    d = df[df.nome_componente == disc]
                    notas.append(d.loc[d.index, 'nota_discente_componente'].get_values()[0])
            if sum(notas) != 0:
                self.medias.append(sum(notas)/len(notas))
            else:
                self.medias.append(sum(notas))
    
    def remove_duplicatas(self):
        # Removendo duplicatas, deixando apenas uma linha por aluno
        self.data = self.data.drop_duplicates(subset='id', keep='first', inplace=False)
    
    def cria_classes(self):
        # Gera as classes para o aprendizado supervisionado

        idade_in  = list(self.data.idade_ingresso_universidade)
        idade_out = list(self.data.idade_colacao_grau)

        for i in range(len(self.data)):
            if (idade_out[i] - idade_in[i]) > 4:
                self.classes.append('NÃO')
            else:
                self.classes.append('SIM')
                
    def removendo_colunas_e_valores_nan(self):
        # Removendo colunas que não são mais úteis para o problema

        self.data = self.data.drop(['id', 'situacao_discente', 'motivo_inativacao', 'data_inativacao', 'data_colacao_grau', 'semestre_cursado', 'codigo_componente', 'nome_componente', 'situacao_discente_componente'], axis=1)
        self.data = self.data.fillna(0)
        
```

## Implementação da Rede Neural

```
import numpy as np
import pandas as pd
import math, random

def read_csv(csv_filename, target_name="target", normalize=False):

    # Leitura do arquivo csv
    df = pd.read_csv(csv_filename, delimiter="\t", dtype={target_name: np.str})

    # Verifica se as classes existem
    if list(df.columns.values).count(target_name) != 1:
        raise Exception("Necessita de exatamente 1 '{}' em {}".format(target_name, csv_filename))

    # Cria um mapeamento com as classes
    map = {}
    targets_unique = sorted(list(set(df[target_name].values)))
    for i, target in enumerate(targets_unique):
        map[target] = i

    def class2idx(y_, map):
        if y_ in map.keys(): return map[y_]
        else: raise Exception("Chave inválida!")

    # Captura as features e as classes
    X = df.drop([target_name], axis=1).values
    y = np.vectorize(class2idx)(df[target_name], map)
    n_classes = len(map.keys())

    # Verifica os tamanho dos dados
    if X.shape[0] != y.shape[0]:
        raise Exception("X.shape = {} e y.shape = {} estão inconsistentes!".format(X.shape, y.shape))

    # Normalize (optional)
    if normalize: X = (X - X.mean(axis=0)) / X.std(axis=0)

    return X, y, n_classes

def crossval_folds(N, n_folds, seed=1):
    np.random.seed(seed)
    idx_all_permute = np.random.permutation(N)
    N_fold = int(N/n_folds)
    idx_folds = []
    for i in range(n_folds):
        start = i*N_fold
        end = min([(i+1)*N_fold, N])
        idx_folds.append(idx_all_permute[start:end])
    return idx_folds
```

```
class NeuralNetwork:

    def __init__(self, input_dim=None, output_dim=None, hidden_layers=None, seed=1):
        if (input_dim is None) or (output_dim is None) or (hidden_layers is None):
            raise Exception("Argumentos inválidos!")
        
        self.input_dim = input_dim                     # Número de nós da camada de entrada
        self.output_dim = output_dim                   # Número de nós da camada de saída
        self.hidden_layers = hidden_layers             # Número de nós de cada camada intermediária
        self.network = self._build_network(seed=seed)  # Rede neural

    # Treinamento da rede neural
    def train(self, X, y, eta=0.5, n_epochs=200):
        for epoch in range(n_epochs):
            for (x_, y_) in zip(X, y):
                self._forward_pass(x_)                              # Etapa farward     (atualiza node["output"])
                yhot_ = self._one_hot_encoding(y_, self.output_dim) # One-hot target
                self._backward_pass(yhot_)                          # Etapa backward    (atualiza node["delta"])
                self._update_weights(x_, eta)                       # Atualiza os pesos (atualiza node["weight"]), atualização baseada na taxa de erro

    # Predict using argmax of logits
    def predict(self, X):
        ypred = np.array([np.argmax(self._forward_pass(x_)) for x_ in X], dtype=np.int)
        return ypred

    # Constrói rede neural totalmente conectada (sem termos de viés)
    def _build_network(self, seed=1):
        random.seed(seed)

        # Cria uma única camada totalmente conectada
        def _layer(input_dim, output_dim):
            layer = [] # Lista de neurônios
            
            for i in range(output_dim):
                weights = [random.random() for _ in range(input_dim)] # Uma amostra de pesos gerados aleatoriamente N(0,1)
                
                node = {"weights": weights,                           # Neurônio que contém os seus pesos
                        "output": None,
                        "delta": None}
                
                layer.append(node) # Camada (ou lista) está guardando cada neurônio com seus respectivos pesos
            
            return layer

        network = [] # Pilha de camadas (entrada -> intermediárias -> saída)
        
        # Se o número de neurônio for zero, então cria uma única camada
        if len(self.hidden_layers) == 0:
            network.append(_layer(self.input_dim, self.output_dim))
        
        else:
            # Caso contrário cria um número m de camadas com um número n de nós
            network.append(_layer(self.input_dim, self.hidden_layers[0])) # Camada de entrada
            
            for i in range(1, len(self.hidden_layers)):
                network.append(_layer(self.hidden_layers[i-1], self.hidden_layers[i])) # Camadas intermediárias
            
            network.append(_layer(self.hidden_layers[-1], self.output_dim)) # Camada de saída

        return network

    # Etada forward (atualiza node['output'])
    def _forward_pass(self, x):
        transfer = self._sigmoid # Função de ativação
        x_in = x # Dados de treinamento
        
        # Para cada camada...
        for layer in self.network:
            x_out = []
            
            # Para cada neurônio da camada 1,2,...,n.
            for node in layer:
                node['output'] = transfer(self._dotprod(node['weights'], x_in)) # Aplica a função de ativação sobre os pesos e os dados de entrada
                x_out.append(node['output'])
            x_in = x_out # Define a saída como próxima entrada
        
        return x_in

    # Etapa backward (atualiza node['delta'], perda L2 é assumida)
    def _backward_pass(self, yhot):
        transfer_derivative = self._sigmoid_derivative # sig' = f(sig) derivada da função
        
        n_layers = len(self.network)
        for i in reversed(range(n_layers)): # Percorre as camadas de modo inverso
            if i == n_layers - 1:
                # Diferença entre logits e one-hot target
                for j, node in enumerate(self.network[i]):
                    err = node['output'] - yhot[j]
                    node['delta'] = err * transfer_derivative(node['output'])
            else:
                # Soma ponderada dos deltas da camada superior
                for j, node in enumerate(self.network[i]):
                    err = sum([node_['weights'][j] * node_['delta'] for node_ in self.network[i+1]])
                    node['delta'] = err * transfer_derivative(node['output'])

    # Atualiza os pesos (atualiza node['weight'])
    def _update_weights(self, x, eta):
        for i, layer in enumerate(self.network):
            # Captura os valores de entrada
            if i == 0: inputs = x
            else: inputs = [node_['output'] for node_ in self.network[i-1]]
            
            # A otimização de peso é a descida de gradiente por meio da regra delta
            for node in layer:
                for j, input in enumerate(inputs):
                    # dw = - learning_rate * (error * transfer') * input
                    node['weights'][j] += - eta * node['delta'] * input

    # Dot product
    def _dotprod(self, a, b):
        return sum([a_ * b_ for (a_, b_) in zip(a, b)])

    # Sigmoid (Função de ativação)
    def _sigmoid(self, x):
        return 1.0/(1.0+math.exp(-x))

    # Derivada da função Sigmoid
    def _sigmoid_derivative(self, sigmoid):
        return sigmoid*(1.0-sigmoid)

    # One-hot encoding
    def _one_hot_encoding(self, idx, output_dim):
        x = np.zeros(output_dim, dtype=np.int)
        x[idx] = 1
        return x
```
